# Handle Missing Values
We can handle the missing values in our data set through these 2 methods
1. Imputation: In this method we can fill the missing the values by statistically calculating the existing data. Either taking Mean, Median or Mode based on the dataset values. If more distributed towards a specific take median

2. Dropping: Or we can drop the data

    Note: Not a recommended way for handling the data. If the data set is large we can drop some data but preferred to handle them statistically

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Imputation

```python
# load dataset
dataset = pd.read_csv('/content/Placement_Dataset.csv')
dataset
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
# count how many missing values
dataset.isnull().sum()
```

```python
# analyze the distribution of data in the salary
fig, axs = plt.subplots(figsize=(8,8))
sns.histplot(dataset.salary, color='red', kde=True, stat="density")

#here we can see the distribution is quite large on the left so taking mean wont be the best choice, instead we can take median
```

```python
# Taking Median
dataset['salary'].fillna(dataset['salary'].median(), inplace=True)
```

```python
dataset.isnull().sum()
```

# Dropping

```python
salary_dataset = pd.read_csv('/content/Placement_Dataset.csv')
salary_dataset.head()
```

```python
salary_dataset.isnull().sum()
```

```python
# drop the missing values
salary_dataset = salary_dataset.dropna(how='any')
```

```python
salary_dataset.isnull().sum()
```
