# Label Encoding
Label Encoding is a technique used in ML in which labels of categorical features are assigned unique integer values.

```python
import pandas as pd;
from sklearn.preprocessing import LabelEncoder;
```

```python
# load the dataset
cancer_dataset = pd.read_csv('/content/breast_cancer_data.csv');
cancer_dataset.head()
```

```python
# finding the count of labels of diagnosis feature
cancer_dataset['diagnosis'].value_counts()
```

```python
# create an instance of LabelEncoder
label_encoder = LabelEncoder();
```

```python
# fit- goes through the categorical features and identifies all the unique categories and then assign unique identifier to it
# transform - applies the learned parameters from the fit() method.

labels = label_encoder.fit_transform(cancer_dataset.diagnosis)
print(labels)
```

```python
# the learned mapping from the fit() method is stored in the classes_ attribute, set in alphabetic order.
print(label_encoder.classes_)
```

```python
# Now we will append the labels to the dataframe
cancer_dataset['target'] = labels;

cancer_dataset.head()
```
